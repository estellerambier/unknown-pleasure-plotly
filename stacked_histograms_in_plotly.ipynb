```python
%load_ext autoreload
%autoreload 2

import numpy as np
import plotly.graph_objects as go
```

```python
def satifying_stacked_histograms_plotly(nb_lines: int, nb_points: int, bins: int, x_min: int, x_max: int):
    fig = go.Figure()

    for i in range(nb_lines):
        # TODO random number of pikes 1 to 3
        # TODO random mu sigma
        mu1, sigma1 = 50, 100
        mu2, sigma2 = 80, 200
        X1 = np.random.normal(mu1, sigma1, nb_points)
        X2 = np.random.normal(mu2, sigma2, nb_points)
        X = np.concatenate([X1, X2])
        X = np.append(X, [x_min, x_max])
        density = np.histogram(X, bins=bins, density=True)
        fig.add_trace(
            go.Scatter(
                y=density[0] + i / 400,
                x=density[1],
                line=dict(color="white", width=0.5),
                showlegend=False,
                fill="tonexty",
                fillcolor="black",
            )
        )
    fig.update_layout(
        paper_bgcolor="black",
        plot_bgcolor="black",
        xaxis=dict(visible=False),
        yaxis=dict(visible=False),
        width=450,
        height=600,
    )
    return fig
```

```python
fig = satifying_stacked_histograms_plotly(80, 400, 100, -950, 950)
fig.show()
```
